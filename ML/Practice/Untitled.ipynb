```python
import pandas as pd
df = pd.read_csv('Bank_Customer_Dataset.csv')
df.head(5)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
df['country'] = encode.fit_transform(df['country'])
df['gender'] = encode.fit_transform(df['gender'])
```

```python
df.head()
```

```python
df.info()
```

```python
X_col = df.columns.tolist()[1:11]
Y_col = df.columns.tolist()[-1]
```

```python
print(X_col)
print(Y_col)
```

```python
X = df[X_col].values
Y = df[Y_col].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
classifier = Sequential()
```

```python
#Layer 1
classifier.add(Dense(6, activation = 'relu', input_shape = (X_train.shape[1], )))
classifier.add(Dropout(rate = 0.8))
```

```python
# Layer 2
classifier.add(Dense(6, activation = 'relu'))
classifier.add(Dropout(rate = 0.6))
```

```python
#O/P layer
classifier.add(Dense(6, activation = 'sigmoid'))
classifier.add(Dense(1, activation = 'sigmoid'))
```

```python
classifier.summary()
```

```python
#weights
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#Fitting the Neural Network
history = classifier.fit(X_train, y_train, batch_size = 32, epochs = 200, validation_split = 0.1)
```

```python
y_pred = classifier.predict(X_test)
print(y_pred[:5])
```

```python
y_pred = (y_pred > 0.3).astype(int)
print(y_pred)
```

```python
#Making the Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
print(((cm[0][0] + cm[1][1])* 100) / len(y_test), '% of data was classified correctly')
```
