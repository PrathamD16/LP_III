```python
curr_x = 2
rate = 0.01
previous_step_size = 1
precision = 0.000001
max_iters = 1000
iters = 0
eqtn = lambda x : 2 * (x+3)
```

```python
while previous_step_size > precision and iters < max_iters:
    pre_x = curr_x
    curr_x -= rate * eqtn(pre_x)
    previous_step_size = abs(pre_x - curr_x)
    print(previous_step_size)
    iters += 1
print(curr_x)
```
