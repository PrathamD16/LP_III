```python
import pandas as pd
df = pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.info()
```

```python
df = df.head(800)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
df['Gender'] = encode.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
import numpy as np
Xcol = df.columns.tolist()[5:13]
Xcol.append(df.columns.to_list()[3])
Ycol = df.columns.tolist()[-1]
print(Xcol)
print(Ycol)
```

```python
X = df[Xcol].values
Y = df[Ycol].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state = 48)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
classifier = Sequential()
```

```python
#Input layer
classifier.add(Dense(64, activation='relu', input_shape=(x_train.shape[1], )))
```

```python
#second layer
classifier.add(Dropout(0.5))
```

```python
#third layer
classifier.add(Dense(8, activation='sigmoid'))
classifier.add(Dropout(0.5))
```

```python
classifier.add(Dense(1, activation='sigmoid'))
```

```python
classifier.summary()
```

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.fit(x_train, y_train, epochs=100)
```

```python
y = classifier.predict(x_test)
print(y)
```

```python
y_predict = [0 if val < 0.5 else 0 for val in y]
print(y_predict[0:5])
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix
cm(y_test, y_predict)
```

```python
print(accuracy_score(y_test, y_predict))
```
